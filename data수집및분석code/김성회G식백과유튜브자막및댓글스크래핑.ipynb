### 유튜브 스크래핑

##### a. Youbute Subtitle

```python
# https://www.youtube.com/watch?v=3VLP69gYhJw
# https://www.youtube.com/watch?v=y75GvVjbNmE
from youtube_transcript_api import YouTubeTranscriptApi
video_id = 'y75GvVjbNmE'
YouTubeTranscriptApi.get_transcript(video_id,languages=['ko']) #영어는 'en', 중국어는 'zh'
```

```python
subtitle = pd.DataFrame(YouTubeTranscriptApi.get_transcript(video_id,languages=['ko']))
subtitle
```

```python
# 엑셀로 저장
subtitle.to_excel(f'{video_id}_subtitle.xlsx')
```

##### b. Youtube Comments

```python
# https://www.youtube.com/watch?v=3VLP69gYhJw
# https://www.youtube.com/watch?v=y75GvVjbNmE 
!youtube-comment-downloader --youtubeid y75GvVjbNmE --output y75GvVjbNmE.json  # json 파일로 다운로드
```

```python
# 다운 받은 json 파일 확인
!ls
```

```python
yt_comments = pd.read_json('y75GvVjbNmE.json',lines=True)
yt_comments
```

```python
# 엑셀로 저장
yt_comments.to_excel(f'{video_id}_comments.xlsx')
```
